```python
import pandas as pd
import numpy as np
```

```python
# Import IMDB data
titles_basics = pd.read_table("title.basics.tsv", 
                              dtype={"tconst": object,
                                     "titleType": object,
                                     "primaryTitle": object,
                                     "originalTitle": object,
                                     "isAdult": int,
                                     "startYear": object,
                                     "endYear": object,
                                     "runtimeMinutes": object,
                                     "genres": object
                                    }
                             )
```

```python
titles_basics
```

```python
titles_basics = titles_basics[titles_basics.titleType == 'movie']
```

```python
titles_basics
```

```python
titles_basics = titles_basics[titles_basics.genres.str.contains('Comedy')]
```

```python
titles_basics
```

```python
# Get rid of porn films
titles_basics = titles_basics[titles_basics.isAdult == 0]
```

```python
titles_basics
```

```python
titles_basics = titles_basics[pd.to_numeric(titles_basics.startYear, errors='coerce').notnull()]
```

```python
titles_basics
```

```python
titles_basics = titles_basics.drop(columns=['endYear', 'isAdult'])
```

```python
titles_basics
```

```python
titles_basics = titles_basics.drop(columns=['titleType', 'runtimeMinutes', 'genres'])
```

```python
titles_basics
```

```python
# Get rid of some foreign language titles.
# Foreign language is not specified in this dataset, but primary and original titles only differ when the film is foreign language
print(titles_basics['primaryTitle'] == titles_basics['originalTitle'])
```

```python
titles_comedy = titles_basics[titles_basics['primaryTitle'] == titles_basics['originalTitle']]
```

```python
titles_comedy
```

```python
titles_comedy.to_csv('titles_comedy.csv')
```

```python
# limit titles_principles only to titles from titles_comedy
titles_principals = pd.read_table('title.principals.tsv',
                           dtype={'tconst': object,
                                  'ordering': int,
                                  'nconst': object,
                                  'category': object,
                                  'job': object,
                                  'characters': object
                                 }
                          )
```

```python
titles_principals
```

```python
comedies_principals = pd.merge(titles_comedy, titles_principals, on='tconst', how='left')
```

```python
comedies_principals
```

```python
comedies_principals = comedies_principals[pd.notnull(comedies_principals['nconst'])]
```

```python
comedies_principals
```

```python
titles_akas= pd.read_table('title.akas.tsv',
                           dtype={'titleId': object,
                                  'ordering': object,
                                  'title': object,
                                  'region': object,
                                  'language': object,
                                  'types': object,
                                  'attributes': object,
                                  'isOriginalTitle': object
                                 }
                          )
```

```python
titles_US = titles_akas[titles_akas['region'] == 'US']
```

```python
titles_US = titles_US[['titleId']]
```

```python
titles_US.columns = ['tconst']
```

```python
US_comedies = pd.merge(titles_US, comedies_principals, on='tconst', how='inner')
```

```python
US_comedies = US_comedies.drop_duplicates()
```

```python
names_basics = pd.read_csv('name.basics.tsv', sep='\t',
                           dtype={'nconst': object,
                                  'primaryName': object,
                                  'birthYear': object,
                                  'deathYear': object,
                                  'primaryProfession': object,
                                  'knownForTitles': object
                                 }
                          )
```

```python
names_basics
```

```python
comedies_principals = pd.merge(US_comedies, names_basics, on='nconst', how='inner')
```

```python
comedies_principals = comedies_principals[comedies_principals.deathYear.str.contains('N')]
```

```python
comedies_principals = comedies_principals[~comedies_principals.birthYear.str.contains('N')]
```

```python
comedies_principals = comedies_principals[pd.to_numeric(comedies_principals.startYear, errors='coerce')]
```

```python
comedies_principals['startYear'] = comedies_principals['startYear'].astype(int)
```

```python
comedies_principals = comedies_principals[comedies_principals.startYear > 1949]
```

```python
comedies_principals = comedies_principals.drop(columns=['job', 'originalTitle', 'deathYear',
                        'primaryProfession', 'knownForTitles'])
```

```python
comedies_principals.to_csv('comedies_principals.csv')
```

```python
comedies_actors = comedies_principals[comedies_principals.category.str.contains('act')]
```

```python
comedies_actors.to_csv('comedies_actors.csv')
```
